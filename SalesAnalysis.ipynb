```python
# Count the occurrences of each product container type
container_counts = data["Product Container"].value_counts()

# Display the result
print(container_counts)
```

```python
# Get the value counts for 'Product Container'
container_counts = data["Product Container"].value_counts()

# Plotting a horizontal bar chart
plt.figure(figsize=(8, 6))
container_counts.plot(kind='barh', color='lightcoral')

# Adding title and labels
plt.title('Frequency of Product Containers')
plt.xlabel('Count')
plt.ylabel('Product Container')
plt.show()
```

```python
# Grouping by 'Region' and summing up 'Shipping Cost'
region_shipping_cost = data.groupby('Region')["Shipping Cost"].sum().reset_index()

# Displaying the result
print(region_shipping_cost)
```

```python
# Create the barplot
plt.figure(figsize=(8, 6))
sns.barplot(x='Region', y='Shipping Cost', data=region_shipping_cost, color='teal')

# Adding labels and title
plt.title('Total Shipping Cost by Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Shipping Cost', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a new column for Year-Month
data['Year-Month'] = data['Order Date'].dt.to_period('M')

# Group by 'Year-Month' and sum up the profits
monthly_profit = data.groupby('Year-Month')['Profit'].sum()

# Display the result
print(monthly_profit)

```

```python
plt.figure(figsize=(12, 6))
Monthly_profit.plot(kind='line', marker='o', color='green')

# Adding title and labels
plt.title('Monthly Profit Trends', fontsize=14)
plt.xlabel('Year-Month')
plt.ylabel('Total Profit')
plt.grid(axis='y', linestyle='--')
plt.xticks(rotation=45)

plt.show()
```

```python
# Create a new column for Year-Month
data['Year-Month'] = data['Order Date'].dt.to_period('M')

# Group by 'Year-Month' and sum up the sales
monthly_sales = data.groupby('Year-Month')['Sales'].sum()

# Display the result
print(monthly_sales)
```

```python
# Plotting the monthly sales trend
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='^', color='blue')

# Adding title and labels
plt.title('Monthly Sales Trends')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.grid(axis='y', linestyle='--')
plt.xticks(rotation=45)
plt.show()
```
